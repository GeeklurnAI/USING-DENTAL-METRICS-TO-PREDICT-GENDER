```python
df.drop(["Sl No","Sample ID","Age"],axis=1,inplace=True)
```

```python
le = LabelEncoder()
df["Gender"]=le.fit_transform(df["Gender"])
```

```python
#x=df.iloc[:,1:11]
#y=df.iloc[:,0]
#x
x=df[["inter canine distance intraoral","intercanine distance casts","right canine width intraoral","right canine width casts","left canine width intraoral","left canine width casts","right canine index intra oral","right canine index casts","left canine index intraoral","left canine index casts"]]
y=df[["Gender"]]
y
```

```python
plt.figure(figsize=(8,8))
cor = df.corr()
sns.heatmap(cor,annot=True,cmap="coolwarm",fmt=".0%")
plt.show()
```

```python
norm = Normalizer()
norm =Normalizer().fit(x)
norm
```

```python
norm.transform(x)
```

```python
x.corr()
```

```python
plt.figure(figsize=(8,8))
cor = x.corr()
sns.heatmap(cor,annot=True,cmap="coolwarm",fmt=".0%")
plt.show()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25,random_state=42)
X_train.shape,X_test.shape
```

```python
log =LogisticRegression()
log.fit(X_train,Y_train)
pred=log.predict(X_test)
```

```python
print(confusion_matrix(Y_test,pred))
```

```python
print(accuracy_score(Y_test,pred))
```

```python
print(classification_report(Y_test,pred))
```

```python
dt = DecisionTreeClassifier()

dt.fit(X_train, Y_train)

pred = dt.predict(X_test)
```

```python
print(confusion_matrix(Y_test,pred))
```

```python
print(accuracy_score(Y_test,pred))
print(classification_report(Y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
pred=rf.predict(X_test)
```

```python
print(confusion_matrix(Y_test,pred))
print(accuracy_score(Y_test,pred))
print(classification_report(Y_test,pred))
print(roc_curve(Y_test,pred))

```

```python
print(roc_auc_score(Y_test,pred))
```

```python
sns.heatmap(confusion_matrix(Y_test, pred), annot=True)
plt.xlabel('Predicted')
plt.ylabel('Ground Truth')

auc = roc_auc_score(Y_test, pred)
fpr, tpr, thresolds = roc_curve(Y_test,pred)
plt.figure(figsize=(12, 6))
plt.axis('scaled')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title("AUC, ROC Curve")
plt.plot(fpr, tpr)
plt.fill_between(fpr, tpr, facecolor='lightblue')
plt.text(0.95, 0.05, 'AUC = %0.4f' % auc, ha='right', fontsize=12)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```
